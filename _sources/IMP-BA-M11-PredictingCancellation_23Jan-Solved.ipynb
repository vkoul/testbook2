##  <font color= darkred> Data Prep for Modeling </font>

### Convert categorical value to numerical

* **[sklearn: `OneHotEncoder`](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html)**  
* **[Pandas: `get_dummies()`](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html)**

```python
df['market_segment_type'].value_counts().plot(kind = 'bar', backend = 'plotly')
```

```python
df['type_of_meal_plan'].value_counts().plot(kind = 'bar', backend = 'plotly')
```

```python
df['room_type_reserved'].value_counts().plot(kind = 'bar', backend = 'plotly')
```

**Dropping the columns not relevant**

```python
df.drop(columns = ['month_name', 'day_name', 'arrival_complete_date', 'year', 'month', 'day', 'arrival_complete_date'], inplace = True)
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
# df.head()
```

### Train-Test Split

```python
# data prep
from sklearn.model_selection import train_test_split
```

```python
# splitting the data 
X = df.drop(columns = 'booking_status')
y = df['booking_status']
```

```python
# # doing train test split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
print("x_train: ", x_train.shape)
print("y_train: ", y_train.shape)


print("\n")

print("x_test: ", x_test.shape)
print("y_test: ", y_test.shape)

# display(x_train.head())
# display(y_train.head())
```

### Scaling 

```python
# x_train.select_dtypes(['int', 'float']).head()
```

```python
to_be_scaled_cols = ['no_of_adults', 'no_of_children', 'lead_time', 'avg_price_per_room']
```

```python
from sklearn.preprocessing import StandardScaler

# intiate 
scaler = StandardScaler()

# fit and transform 
x_train_scaled = scaler.fit_transform(x_train[to_be_scaled_cols])

# transform 
x_test_scaled = scaler.transform(x_test[to_be_scaled_cols])
```

```python
type(x_test_scaled)
```

### Dummy classifier model using [DummyClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.dummy.DummyClassifier.html)

```python
from sklearn.dummy import DummyClassifier
```

```python
# call the model 
dummy = DummyClassifier()

# fit the model
dummy.fit(x_train, y_train)
```

```python
print("The accuracy score of the model for Train is {:.2f}".format(dummy.score(x_train, y_train)))

print("The accuracy score of the model for Test is {:.2f}".format(dummy.score(x_test, y_test)))
```

### Logistic Regression model using [LogisticRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html)

```python
# call the model 
lr = LogisticRegression()

# fit the model 
lr.fit(x_train, y_train)
```

### k-nn model using [KNeighborsClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html)

### Decision Trees model using [DecisionTreeClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html)

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# call the model
dt = DecisionTreeClassifier()

# fit the model 
dt.fit(x_train, y_train)
```

```python
print("The accuracy score of the model for Train is {:.2f}".format(dt.score(x_train, y_train)))

print("The accuracy score of the model for test is {:.2f}".format(dt.score(x_test, y_test)))
```

#### Feature Importance in DT

```python
# list of column name
feature_names = list(x_train.columns)

# get the importanc from the model
importances = dt.feature_importances_

# placement
indices = np.argsort(importances)

# plotting the figure
plt.figure(figsize = (8, 8))
plt.title("Decision Tree Feature Importance")

plt.barh(range(len(indices)), importances[indices], color="darkred", align="center")

plt.yticks(range(len(indices)), [feature_names[i] for i in indices])
plt.xlabel("Relative Importance")

plt.show()
```

#### Visualize the Tree

```python
from sklearn import tree
```

```python
features = list(x_train.columns)

plt.figure(figsize=(20,20))

tree.plot_tree(dt,feature_names= features,max_depth = 3, filled=True, fontsize= 9, node_ids = True, class_names= True)
plt.show()
```

#### GridSearch CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
# define the parameters
params = {'max_leaf_nodes': list(range(2, 6))
          ,'min_samples_split': [2, 3, 4]
          ,'criterion':['gini','entropy']
          ,'max_depth':list(range(4,10))}

# check the params
params
```

```python
# intialise the GS CV
grid_search_cv = GridSearchCV(DecisionTreeClassifier(random_state=42), params, verbose=1, cv=3)

# fit the model
grid_search_cv.fit(x_train, y_train)
```

```python
grid_search_cv.best_estimator_
```

```python
gs_df = pd.DataFrame(grid_search_cv.cv_results_).sort_values(by = 'rank_test_score')
display(gs_df.head())
```
